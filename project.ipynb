```python
xxx = [[data[cell]['cycles']['10']['Qdlin'], data[cell]['cycles']['100']['Qdlin']] for cell in data.keys()][:1000]
yyy = [[data[cell]['cycles']['10']['Vdlin'], data[cell]['cycles']['100']['Vdlin']] for cell in data.keys()][:1000]

delta_Q = []

for i in range(len(xxx)):
    delta_Q.append(xxx[i][1]-xxx[i][0])

for el in delta_Q:
    plt.plot(el, yyy[0][0])

print(xxx[0][0].shape)
```

## Studying the minima of the deltaQ plot

```python
cycles = [data[cell]['cycle_life'] for cell in data.keys()]
# log_cycles = [np.log(cycle) for cycle in cycles]
print(len(cycles))

min_delta_Q = [abs(min(delta_Q[i])) for i in range(len(delta_Q))]

```

## Studying the surface of the deltaQ plot

```python
sum_delta_Q = [abs(sum(delta_Q[i])) for i in range(len(delta_Q))]


```

```python
plt.subplot(2, 2, 1)  # row 1, col 2 index 1
plt.scatter(min_delta_Q, cycles, c=cycles, cmap='viridis_r')
plt.yscale("log")
plt.xscale("log")
plt.xlabel("min_delta_Q")
plt.ylabel("cycles")
plt.title("Min of the deltaQ plot")

plt.subplot(2, 2, 2)  # index 2
plt.scatter(sum_delta_Q, cycles, c=cycles, cmap='viridis_r')
plt.yscale("log")
plt.xscale("log")
plt.xlabel("min_delta_Q")
plt.ylabel("cycles")
plt.title("surface of the deltaQ plot")


plt.show()
```
